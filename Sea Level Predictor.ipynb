```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress
```

```python
def draw_plot():
    # Read data from file
  df = pd.read_csv("epa-sea-level.csv")


    # Create scatter plot
  fig = plt.figure(figsize=(12, 6))
  ax = fig.add_subplot(1, 1, 1)

  # Plot scatter plot
  sc = ax.scatter(df["Year"], df["CSIRO Adjusted Sea Level"], c=df["CSIRO Adjusted Sea Level"], cmap='Spectral')

  # Add colorbar and labels
  plt.colorbar(sc)
  ax.set_xlabel("Year")
  ax.set_ylabel("CSIRO Adjusted Sea Level")

  # Show plot
  plt.show()

    # Create first line of best fit
  slope, intercept, rvalue, pvalue, stderr = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])

# Create figure and subplot
  fig = plt.figure(figsize=(12, 6))
  ax = fig.add_subplot(1, 1, 1)

# Plot scatter plot
  ax.scatter(df["Year"], df["CSIRO Adjusted Sea Level"], s=20, alpha=0.5, color="blue")

# Plot line of best fit
  x = np.arange(1880, 2051)
  y = slope * x + intercept
  ax.plot(x, y, color="red", linewidth=3)

# Add labels and legend
  ax.set_xlabel("Year")
  ax.set_ylabel("CSIRO Adjusted Sea Level")
  ax.set_title("Sea Level Rise")
  ax.legend(["Line of best fit", "Data"])

# Show plot
  plt.show()

    # Create second line of best fit
  # Filter the data for years after 2000
  df = df[df['Year'] >= 2000]

# Perform linear regression on the filtered data
  slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

# Create a range of years from 2000 to 2050
  years = np.arange(2000, 2051)

# Calculate the sea level rise predictions for the range of years
  predictions = slope * years + intercept

# Create a figure and axis object
  fig, ax = plt.subplots(figsize=(12, 6))

# Add the scatter plot to the axis object
  ax.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], s=20)

# Add the line of best fit to the axis object
  ax.plot(years, predictions, 'r', label='Line of best fit')

# Set the axis labels and title
  ax.set_xlabel('Year')
  ax.set_ylabel('Sea Level (inches)')
  ax.set_title('Sea Level Rise Since 2000')

# Add a legend to the plot
  ax.legend()

# Show the plot
  plt.show()

    # Add labels and title
  ax.set_xlabel("Year")
  ax.set_ylabel("Sea Level (inches)")
  ax.set_title("Rise in Sea Level")
  ax.legend(["Line of best fit", "Data"])
    
    # Save plot and return data for testing (DO NOT MODIFY)
  plt.savefig('sea_level_plot.png')
  return plt.gca()
```

## In usage

```python
df = pd.read_csv("epa-sea-level.csv")

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1)

sc = ax.scatter(df["Year"], df["CSIRO Adjusted Sea Level"],       c=df["CSIRO Adjusted Sea Level"], cmap='Spectral')

plt.colorbar(sc)
ax.set_xlabel("Year")
ax.set_ylabel("CSIRO Adjusted Sea Level")

plt.show()
```

```python
slope, intercept, rvalue, pvalue, stderr = linregress(df["Year"], df["CSIRO Adjusted Sea Level"])

fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1)

ax.scatter(df["Year"], df["CSIRO Adjusted Sea Level"], s=20, alpha=0.5, color="blue")

x = np.arange(1880, 2051)
y = slope * x + intercept
ax.plot(x, y, color="red", linewidth=3)

ax.set_xlabel("Year")
ax.set_ylabel("CSIRO Adjusted Sea Level")
ax.set_title("Sea Level Rise")
ax.legend(["Line of best fit", "Data"])

plt.show()
```

```python
df = df[df['Year'] >= 2000]

slope, intercept, r_value, p_value, std_err = linregress(df['Year'], df['CSIRO Adjusted Sea Level'])

years = np.arange(2000, 2051)

predictions = slope * years + intercept

fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(df['Year'], df['CSIRO Adjusted Sea Level'], s=20)

ax.plot(years, predictions, 'r', label='Line of best fit')

ax.set_xlabel('Year')
ax.set_ylabel('Sea Level (inches)')
ax.set_title('Sea Level Rise Since 2000')

ax.legend()

plt.show()

ax.set_xlabel("Year")
ax.set_ylabel("Sea Level (inches)")
ax.set_title("Rise in Sea Level")
ax.legend(["Line of best fit", "Data"])
```
